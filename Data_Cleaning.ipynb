Let's have a view of the Listed_in

```python
Netflix_cleaning.listed_in
```

 This need to be separated into consituents

```python
Types_List = []
for x, y in Netflix_cleaning.iterrows():
  for item in y['listed_in'].split(','):
    if item.strip() not in Types_List:
      Types_List.append(item.strip())
```

```python
Types_List
```

```python
df = [0]*len(Types_List)
for r, i in Netflix_cleaning.iterrows():
  for item in i['listed_in'].split(','):
    p = Types_List.index(item.strip())
    df[p]+=1
```

```python
df_list = pd.DataFrame({'Type': Types_List, 'Count':df})
```

```python
df_list.sort_values(by = 'Count',ascending=False)
```
